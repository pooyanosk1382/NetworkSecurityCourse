```python
from sympy import Matrix
from itertools import product
```

```python
def decryptHillCipher(cipherText, keySize, alphabet):
    n = keySize
    alphabetSize = len(alphabet)
    inverse = lambda a, m: pow(a, -1, m)
    for key in product(range(alphabetSize), repeat=n**2):
        keyMatrix = Matrix(key).reshape(n, n)
        try:
            inverseKeyMatrix = keyMatrix.inv_mod(alphabetSize)
        except ValueError:
            continue
        plainText = ""
        for i in range(0, len(cipherText), n):
            cipherBlock = Matrix([alphabet.index(c) for c in cipherText[i:i+n]]).reshape(n, 1)
            plainBlock = (inverseKeyMatrix * cipherBlock) % alphabetSize
            plainText += "".join([alphabet[x] for x in plainBlock])
        yield plainText
```

```python
cipherText = "HJHL"
keySize = 2
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
for plainText in decryptHillCipher(cipherText, keySize, alphabet):
    print(plainText)
```
